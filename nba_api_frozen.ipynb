```python
import pandas as pd
import time
from nba_api.stats.endpoints import LeagueGameFinder

seasons = ['2020-21', '2021-22', '2022-23', '2023-24', '2024-25']
frames = []

for season in seasons:
    print(f'Pulling {season}...')
    finder = LeagueGameFinder(
        season_nullable=season,
        season_type_nullable='Regular Season'
    )
    frames.append(finder.get_data_frames()[0])
    time.sleep(2)

raw = pd.concat(frames, ignore_index=True)
print(f'Done. Total rows: {len(raw):,}')
raw.head()
```

```python
# Save and download — this is the frozen CSV
raw.to_csv('nba_games_raw.csv', index=False)
print(f'Saved {len(raw):,} rows to nba_games_raw.csv')

from google.colab import files
files.download('nba_games_raw.csv')
print('Download started — check your Downloads folder.')
```
